```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.load('X.npy')
```

### Affichage de 9 images du dataset

### Gnration d'images avec rotation alatoire

```python
datagen = ImageDataGenerator(
    rotation_range=np.random.random_sample()
)

datagen.fit(X_selected_reshaped)
imgs = datagen.flow(X_selected_reshaped, shuffle=False, batch_size=9)

plt.figure(figsize=(10, 10))
i = 1
for img in imgs:
    plt.subplot(3, 3, i)
    plt.imshow(img[i-1].reshape(64,64))
    i += 1
    if i > 9:
        break
plt.show()
```

### Gnration d'images avec translation alatoire

```python
datagen = ImageDataGenerator(
    width_shift_range=np.random.random_sample(),
    height_shift_range=np.random.random_sample()
)

datagen.fit(X_selected_reshaped)
imgs = datagen.flow(X_selected_reshaped, shuffle=False, batch_size=9)

plt.figure(figsize=(10, 10))
i = 1
for img in imgs:
    plt.subplot(3, 3, i)
    plt.imshow(img[i-1].reshape(64,64))
    i += 1
    if i > 9:
        break
plt.show()
```
